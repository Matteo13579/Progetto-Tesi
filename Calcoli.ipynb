```python
def quartile(df):
    idx_q1 = (0.25 * (len(df) + 1))
    idx_q2 = (0.50 * (len(df) + 1))
    idx_q3 = (0.75 * (len(df) + 1))

    if idx_q1.is_integer():
        Q1 = df[int(idx_q1) - 1]
    else:
        lower_idx = int(idx_q1) - 1
        upper_idx = int(idx_q1)
        Q1 = ((df[lower_idx] + df[upper_idx]) / 2)
    if idx_q2.is_integer():
        Q2 = df[int(idx_q2) - 1]
    else:
        lower_idx = int(idx_q2) - 1
        upper_idx = int(idx_q2)
        Q2 = ((df[lower_idx] + df[upper_idx]) / 2)
    if idx_q3.is_integer():
        Q3 = df[int(idx_q3) - 1]
    else:
        lower_idx = int(idx_q3) - 1
        upper_idx = int(idx_q3)
        Q3 = ((df[lower_idx] + df[upper_idx]) / 2)

    return Q1, Q2, Q3
```

```python
Quartile_df = pd.DataFrame(columns=['Q1', 'Q2', 'Q3'])

for col in df.columns:
    property = df[col].copy().sort_values().reset_index(drop=True)
    Q1, Q2, Q3 = quartile(property)

    Quartile_df.loc[col] = [Q1, Q2, Q3]

Quartile_df
```

```python
Quartile_df["IQR"] = (Quartile_df['Q3'] - Quartile_df['Q1'])
Quartile_df
```

```python
Quartile_df["Min"] = (df.min(axis=0).round(2))
Quartile_df["Max"] = (df.max(axis=0).round(2))
Quartile_df
```

```python
Lower_bound = Quartile_df['Q1'] - 1.5 * Quartile_df['IQR']
Upper_bound = Quartile_df['Q3'] + 1.5 * Quartile_df['IQR']

Quartile_df["Lower Bound"] = Lower_bound
Quartile_df["Upper Bound"] = Upper_bound

Quartile_df
```

```python
from IPython.display import display as ds

temp_df = pd.DataFrame()
df_copy = df.copy().drop(columns=['quality']) # Drop non-numerical columns



```

```python
num_cols = df_copy.select_dtypes(include=['float64', 'int64']).columns.tolist() # Get numerical columns
if len(num_cols) > 0:
    first_time = True
    outlier_cols = []
    for col in num_cols:
        q1 = df_copy[col].quantile(0.25) # Get the first quartile
        q3 = df_copy[col].quantile(0.75) # Get the third quartile
        iqr = q3 - q1 # Get the interquartile range
        lower_bound = q1 - 1.5 * iqr # Get the lower bound
        upper_bound = q3 + 1.5 * iqr # Get the upper bound
        outliers = df_copy[(df_copy[col] < lower_bound) | (df_copy[col] > upper_bound)][col] # Get the outliers

        if not outliers.empty:
            temp = pd.DataFrame(outliers)
            temp.columns = [col]
            temp_df = pd.concat([temp_df, temp], axis=1)

temp_df.fillna(0, inplace=True)

temp_df["outlier_count"] = (temp_df != 0).sum(axis=1)

print(temp_df["outlier_count"].value_counts()), temp_df["outlier_count"].__len__()
```
